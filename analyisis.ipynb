```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import utils


G = utils.read_pajek(data_folder="data",filename="correct_epo_collaboration_graph.net")
#G = nx.read_edgelist("data/correct_epo_collaboration_graph.net")
utils.info(G)

connected_components = list(nx.connected_components(G))
number_of_components = len(connected_components)

print(f"Number of connected components: {number_of_components}")6
```

```python
#degree distirbution
degree_sequence1 = [sublist[1] for sublist in G.degree()]
degree_counts1 = np.bincount(degree_sequence1)
pk1 = degree_counts1 / len(degree_sequence1)

plt.figure(figsize=(10, 6))
plt.loglog(np.nonzero(pk1)[0], pk1[np.nonzero(pk1)[0]], 'ro', label='Degree Distribution1')

plt.xlabel('Degree (k)')
plt.ylabel('Probability (p_k)')
plt.title('Degree Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def sort_and_print(nodes):
    nodes = sorted(nodes.items(), key=lambda x: x[1], reverse=True)    
    for i in range(0,10):        
        print(G.nodes[nodes[i][0]]['label']," with centrality score of: ",nodes[i][1])
    #for node in nodes:
    #    print("Value of node ",node)
    #    print(G.nodes(node[0]))

    #print(nodes)

# 1. Degree Centrality
degree_centrality = nx.degree_centrality(G)
print("10 nodes with biggest degree centrality measure:\n")
sort_and_print(degree_centrality)
#print("Degree Centrality:", degree_centrality)


# 3. Closeness Centrality
closeness_centrality = nx.closeness_centrality(G)
print("\n10 nodes with biggest closeness centrality measure:\n")
sort_and_print(closeness_centrality)


# 4. PageRank
pagerank = nx.pagerank(G)
print("\n10 nodes with biggest pagerank measure:\n")
sort_and_print(pagerank)

# 2. Betweenness Centrality
betweenness_centrality = nx.betweenness_centrality(G)
print("\n10 nodes with biggest betweenness centrality measure")
sort_and_print(betweenness_centrality)
#print("Betweenness Centrality:", betweenness_centrality)
```
